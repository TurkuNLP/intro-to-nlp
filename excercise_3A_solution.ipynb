```python
from sklearn.metrics import cohen_kappa_score
```

```python
# wget the data
!wget --quiet https://raw.githubusercontent.com/TurkuNLP/intro-to-nlp/refs/heads/master/Data/emotion-sample.tsv
```

```python
text = []
annotations = []

# Read the data
with open("emotion-sample.tsv", "r") as f:
  for line in f:
    line = line.strip("\n") # remember to remove trailing newline character
    if line:
      t, l = line.split("\t") # text and label are separated by tab "\t"
      text.append(t)
      # convert label to integer
      annotations.append(int(l))
```

```python
# Create id to label mapping and print it for reference while annotating
id2label = {0: "sadness", 1: "joy", 2: "love", 3: "anger", 4: "fear", 5: "surprise"}
for id, label in id2label.items():
  print(id, label)
```

```python
my_annotations = []
for t in text:
  print(t)
  ann = int(input("Give annotation: ")) # Convert input to integer
  my_annotations.append(ann)
```

```python
# Let's see that we got what we expected
my_annotations
```

```python
# Calcualate Cohen's Kappa using cohen_kappa_score from sklearn
kappa = cohen_kappa_score(my_annotations, annotations)
print(kappa)
```

So what is actually a good Kappa score? There are no official guidelines or consensus, but roughly:
- less than 0.4: Bad
- 0.4-0.6: Okay
- 0.6-0.8:  Good
- more than 0.8: Great

```python
# This is the extra part for those interested


# Hmmm, not the greatest agreement.
# Let's see where I disagreed with the original labels

# First, let's convert our integer labels to actual human-readable labels
def convert_to_text_labels(labels):
  text_labels = []
  for l in labels:
    text_labels.append(id2label[l])

  return text_labels

my_labels = convert_to_text_labels(my_annotations)
original_labels = convert_to_text_labels(annotations)
```

```python
for t, m, o in zip(text, my_labels, original_labels):
  print(t)
  print("My label:", m)
  print("Original label:", o)
  print("")
```

```python
# Let's put our labels into a confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Use the integer values to get correct matrix
cm = confusion_matrix(annotations, my_annotations)
print(cm)
```

```python
# Alternatively, use string labels
#but remember to give the "labels" argument with the labels in correct order

cm = confusion_matrix(original_labels, my_labels, labels=list(id2label.values()))
print(cm)
```

```python
# Visualise the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(id2label.values()))
disp.plot()
```
