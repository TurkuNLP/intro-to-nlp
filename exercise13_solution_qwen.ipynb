# Exercise 13: Text generation

This example solution uses Qwen/Qwen3.5-0.8B (instruction fine-tuned) and Qwen/Qwen3.5-0.8B-base (base LM) models without quantization.


```python
# using Qwen/Qwen3.5-0.8B models requires newer transformers version
!pip install --upgrade transformers
```

```python
from transformers import pipeline
```

```python
MODEL_NAME = 'Qwen/Qwen3.5-0.8B'

pipe = pipeline("text-generation", model=MODEL_NAME)
```

```python
task_prompts = ["What is the capital of Finland?",
                "What is the sentiment in the sentence: I enjoyed the movie, the plot was great.",
                "List all named entities (PERSON, LOCATION, ORGANIZATION) from the sentence: The president of Finland is Alexander Stubb."]
conversation_template = [{"role": "user", "content": ""}]

for task_prompt in task_prompts:
  conversation_template[0]["content"] = task_prompt
  output = pipe(conversation_template, max_new_tokens=50)
  for c in output[0]["generated_text"]:
    print(f"{c['role']}: {c['content']}")
  print("\n----------------\n")
```

Now restart the session to free memory if running big models. Then try the same with base LM.

```python
from transformers import pipeline
```

```python
MODEL_NAME = 'Qwen/Qwen3.5-0.8B-base'

pipe = pipeline("text-generation", model=MODEL_NAME)
```

```python
task_prompts = ["What is the capital of Finland?",
                "What is the sentiment in the sentence: I enjoyed the movie, the plot was great.",
                "List all named entities (PERSON, LOCATION, ORGANIZATION) from the sentence: The president of Finland is Alexander Stubb."]


for prompt in task_prompts:
  output = pipe(prompt, max_new_tokens=200)
  print(output[0]["generated_text"])
  print("\n--------------------\n")

```

Based on the output, the model training has definitely included more than just pre-training phrase (as often in the base models) since the model starts to "think".

This model appears to be fine-tuned with reasoning data (supervised language modeling where the correct continuation includes reasoning before giving the final output).

When reading the MODEL CARD (https://huggingface.co/Qwen/Qwen3.5-0.8B-Base) it actually says `Training Stage: Pre-training & Post-training` so this is not a standard base model.

At least partly from this reason, the outputs look very similar for both models.
